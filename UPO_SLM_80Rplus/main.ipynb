```python
from SLM_UPOLabs import *
slm = SLM_UP()
slm.Open_window(1) #打开slm窗口
resX, resY = slm.Get_size(1) #得到slm的分辨率
resX, resY
```

```python
X,Y = np.meshgrid(np.linspace(0,resX-1,resX),np.linspace(0,resY-1,resY))
testIMG = np.round((2**8-1)*(0.5+0.5*np.sin(2*np.pi*X/50))).astype('uint8').T
testIMG.shape#图像大小
slm.Disp_Data_Int(testIMG, 1, 8) #int数据显示
```

```python
random_array = np.random.uniform(0, 255, size=(resX, resY))
slm.Disp_Data_Double(random_array, 1, 8) #double数据显示
```
